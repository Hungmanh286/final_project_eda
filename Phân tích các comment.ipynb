```python
res=[]
x = []
for post in post_list:
    if len(post['comments_full']) > 0:
      print(post['comments_full'])
      x.append(post['comments_full'])
      cmt = '\n'.join([comment['comment_text'] for comment in post['comments_full'] ])
      res.append(cmt)
len(res)
res
```
