```python
post_df_full1['reactions'] = post_df_full1['reactions'].apply(lambda x : dict(eval(x)) )
post_df_full_reactions = post_df_full1['reactions'].apply(pd.Series)
post_df_full_reactions
```

```python
post_df_full_with_reactions = pd.concat([post_df_full, post_df_full_reactions], axis=1).drop('reactions', axis=1)
post_df_full_with_reactions[['post_text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments','reaction_count']]
```

# Thay c√°c gi√° tr·ªã l√† nan b·∫±ng gi√° tr·ªã c√≥ t·∫ßn su·∫•t nhi·ªÅu nh·∫•t trong c·ªôt


```python
for i in range(0, len(post_df_full_with_reactions['th√≠ch'].values), 1):
    if pd.isna(post_df_full_with_reactions['th√≠ch'].values[i]):
        post_df_full_with_reactions['th√≠ch'].values[i] = post_df_full_with_reactions['th√≠ch'].mode()[0]

    if pd.isna(post_df_full_with_reactions['y√™u th√≠ch'].values[i]):
        post_df_full_with_reactions['y√™u th√≠ch'].values[i] =post_df_full_with_reactions['y√™u th√≠ch'].mode()[0]


    if pd.isna(post_df_full_with_reactions['haha'].values[i]):
        post_df_full_with_reactions['haha'].values[i] = post_df_full_with_reactions['haha'].mode()[0]


    if pd.isna(post_df_full_with_reactions['wow'].values[i]):
        post_df_full_with_reactions['wow'].values[i] = post_df_full_with_reactions['wow'].mode()[0]


    if pd.isna(post_df_full_with_reactions['bu·ªìn'].values[i]):
        post_df_full_with_reactions['bu·ªìn'].values[i] =  post_df_full_with_reactions['bu·ªìn'].mode()[0]


    if pd.isna(post_df_full_with_reactions['ph·∫´n n·ªô'].values[i]):
        post_df_full_with_reactions['ph·∫´n n·ªô'].values[i] = post_df_full_with_reactions['ph·∫´n n·ªô'].mode()[0]


    if pd.isna(post_df_full_with_reactions['th∆∞∆°ng th∆∞∆°ng'].values[i]):
        post_df_full_with_reactions['th∆∞∆°ng th∆∞∆°ng'].values[i] = post_df_full_with_reactions['th∆∞∆°ng th∆∞∆°ng'].mode()[0]


    if pd.isna(post_df_full_with_reactions['shares'].values[i]):
        post_df_full_with_reactions['shares'].values[i] = post_df_full_with_reactions['shares'].mode()[0]


    if pd.isna(post_df_full_with_reactions['comments'].values[i]):
        post_df_full_with_reactions['comments'].values[i] = post_df_full_with_reactions['comments'].mode()[0]

post_df_full_with_reactions['reaction_count'] = post_df_full_with_reactions[['th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments']].sum(axis=1)

post_df_full_with_reactions[['post_text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments','reaction_count']]
```

```python
print(post_df_full_with_reactions['th√≠ch'].mean())
print(post_df_full_with_reactions['th√≠ch'].min())
print(post_df_full_with_reactions['th√≠ch'].max())
```

Bi·ªÉu ƒë·ªì th·ªÉ hi·ªán l∆∞·ª£ng t∆∞∆°ng t√°c c·ªßa t·ª´ng c·∫£m x√∫c theo th·ªùi gian

```python
reactions = ['th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng', 'shares', 'comments', 'reaction_count']
num_rows = 5
num_cols = len(reactions) // num_rows + (len(reactions) % num_rows > 0)
fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(28, 15))
axs = axs.flatten()
for i, reaction in enumerate(reactions):
    axs[i].plot(post_df_full_with_reactions['time'], post_df_full_with_reactions[reaction], label=reaction, marker="o")
    axs[i].legend()
    axs[i].xaxis.set_major_formatter(DateFormatter('%d-%m-%y'))
    axs[i].set_title(f"{reaction}")
plt.tight_layout()
plt.show()
```
