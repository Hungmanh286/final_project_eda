```python
from facebook_scraper import get_posts
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

FANPAGE_LINK ="iVIVU"
FOLDER_PATH = "/content/drive/MyDrive/Colab Notebooks/crawl_facebook/"
COOKIE_PATH = "/content/drive/MyDrive/Colab Notebooks/crawl_facebook/www.facebook.com_cookies.txt"
PAGES_NUMBER = 10 # số trang cần cào
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=10, cookies=COOKIE_PATH):
    post_list.append(post)
```
